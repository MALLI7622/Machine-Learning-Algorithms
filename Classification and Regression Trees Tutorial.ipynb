```python
def CART( X1, X2, Y):
    
    epochs = 10
    
    for i in range(epochs):
        
        
        first_spilt = X1[i]
    
        print(first_spilt)
        group = []
        
        predict = []
        
        for i in X1:
            if i >= first_spilt: 
                group.append("Right")
                predict.append(1)
            else:
                group.append("Left")
                predict.append(0)
            
            
        count_r0 = 0
        count_r1 = 0
        count_l0 = 0
        count_l1 = 0
    
        count_right = group.count("Right")
        count_left = group.count("Left")
    
        try:
            
            for i, j in zip(Y, group):

                if j == 'Right' and i == 0:
                    count_r0 += 1
                elif j == 'Right' and i == 1:
                    count_r1 += 1
                elif j == 'Left' and i == 1:
                    count_l1 += 1
                else:
                    count_l1 += 1

            print(count_right, count_left, count_r0, count_r1, count_l0, count_l1)  

            a = ((1 - (count_r0 ** 2 / count_right + count_r1 ** 2 / count_right)) * (count_r0+count_r1) / len(Y))

            b = ((1 - (count_l0 ** 2 / count_left + count_l1 ** 2 / count_left)) * (count_l0+count_l1) / len(Y))

            print(a+b)

            for i, j, k in zip(Y, predict, group):
                print(i, j, k)

        except ZeroDivisionError:
            pass
```

```python
CART(X1, X2, Y)
```
