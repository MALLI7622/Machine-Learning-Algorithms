# Text Mercato Assignment

```python
## Importing all the libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
## Opening the directory
pics = os.listdir('pics/')
```

```python
pics = pics[1:]
len(pics)
```

```python
pics
```

## Plotting all the images

```python
plt.figure(figsize=(25,25))
for i in range(20):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(True)
    img = cv2.imread('pics/'+pics[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img, cmap='Greys_r')
    plt.xlabel('img'+str(i+1))
    plt.title(img.shape)
plt.show()
```

## Data Preparation

Images contain different sizes, so changing them into equal dimensions of all images, by doing this I had cropped the changed the width of the images into equal sizes by taking minimum sized image width

```python
data = []
plt.figure(figsize=(25,25))
for i in range(20):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(True)
    img = cv2.imread('pics/'+pics[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img[:, :323,:]
    im = img.copy()
    im = im.flatten()
    data.append(im)
    plt.imshow(img, cmap='Greys_r')
    plt.xlabel('img'+str(i+1))
    plt.title(img.shape)
plt.show()
```

## Training

This is an unsupervised problem, So I am using KMeans algorithm to solve this problem

```python
k = 5
kmeans = KMeans(n_clusters = k)
y_pred = kmeans.fit_predict(data)
```

```python
y_pred
```

## After Training images and corresponding product labels

```python
plt.figure(figsize=(25,25))
for i in range(20):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(True)
    img = cv2.imread('pics/'+pics[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img, cmap='Greys_r')
    plt.xlabel('img'+str(i+1))
    plt.title("product: "+ str(y_pred[i]))
plt.show()
```
